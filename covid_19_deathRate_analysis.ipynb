```python
import numpy as np
import pandas as pd 
import seaborn as sns 
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')
sns.set_style("white")
```

```python
data = pd.read_csv('/content/covid_19_india.csv')
```

```python
data.head()
```

```python
colors1 = ['#e6ccb2','#b08968','#7f5539'] 
colors2 = ['#ede0d4','#e6ccb2','#ddb892','#b08968','#7f5539','#9c6644','#dda15e']
```

```python
sns.set(rc = {'figure.figsize':{25,5}})
plot = sns.countplot(x = 'State/UnionTerritory',data = data)
plot.set_xticklabels(plot.get_xticklabels(),rotation = 40)
plt.show()
```

```python
data.dtypes
```

```python
data["Date"] = pd.to_datetime(data["Date"])
```

```python
ax = plt.subplot(1,2,2)
sns.set(rc = {'figure.figsize':[15,5]})
sns.lineplot(x = "Date",y = 'Cured',data =data,color = "red")

ax = plt.subplot(1,2,1)
sns.set(rc = {'figure.figsize':[15,5]})
sns.lineplot(x = "Date",y = 'Deaths',data =data,color = "red")
```

```python
sns.set(rc = {'figure.figsize':{10,5}})
plot = sns.histplot(data = data['Deaths'])
plt.show()
```

```python
sns.scatterplot(x = 'Date',y = 'Deaths',style = True,palette = True,data = data,color ='green')
plt.show()
```

```python
sns.pairplot(data = data)
plt.show()
```

```python
correlation = data.corr()
sns.heatmap(correlation,annot= True,square = True,cmap = colors2)
plt.show()
```
